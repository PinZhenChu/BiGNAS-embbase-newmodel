```python
import numpy as np
import pandas as pd
from collections import Counter

E_add_source = np.load("E_add_source.npy")
E_add_target = np.load("E_add_target.npy")

# === 統計被加邊次數 ===
item_count = Counter(E_add_target)
user_count = Counter(E_add_source)

# 轉成 DataFrame
item_df = pd.DataFrame(item_count.items(), columns=["item_id", "added_count"]).sort_values("added_count", ascending=False)
user_df = pd.DataFrame(user_count.items(), columns=["user_id", "added_count"]).sort_values("added_count", ascending=False)

print("📈 Top 10 most frequently added items:")
print(item_df.head(10))

print("\n👤 Top 10 users who got most new edges:")
print(user_df.head(10))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(item_df["added_count"], bins=50)
plt.title("Item Added Count Distribution")
plt.xlabel("Number of times item was added")
plt.ylabel("Frequency")
plt.show()
```
